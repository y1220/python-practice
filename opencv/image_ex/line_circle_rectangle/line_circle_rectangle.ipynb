```python
import cv2
import numpy as np
```

```python
img = cv2.imread("data/src/road.jpg")
img_g = cv2.imread("data/src/road.jpg",0)
cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_canny = cv2.Canny(img_g, 300, 450)
cv2.imshow("img", img_canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
lines = cv2.HoughLines(img_canny, 1, np.pi/100, 100) # 100 times show up
# theta width
```

```python
lines # (rho, theta)
```

```python
for i in lines[:]:
    rho = i[0][0]
    theta = i[0][1]
    a = np.cos(theta) 
    b = np.sin(theta)    
    x0 = rho * a
    y0 = rho * b
    x1 = int(x0 + 1000 * (-b)) # starting point
    y1 = int(y0 + 1000 * (a))    
    x2 = int(x0 + 1000 * (-b)) # ending point  
    y2 = int(y0 + 1000 * (a))
    cv2.line(img, (x1, y1), (x2, y2), (255, 0, 0), 1)
cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# detect circle
img2 = cv2.imread("data/src/grapes.jpg")
img2_g = cv2.imread("data/src/grapes.jpg",0)
cv2.imshow("img", img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
circles = cv2.HoughCircles? # see params list
```

```python
circles = cv2.HoughCircles(img2_g, cv2.HOUGH_GRADIENT, dp = 1, minDist=1, param1 = 20, param2 = 35, minRadius = 1, maxRadius = 30)
```

```python
for i in circles[0]:
    cv2.circle(img2, (i[0], i[1]), i[2], (255, 0, 0), 1)
cv2.imshow("img",img2)
cv2.waitKey(0)
cv2.destroyAllWindows()            
```

```python
# detect rectangle
image = cv2.imread('data/src/rec.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
canny = cv2.Canny(gray, 130, 255, 1)

cnts = cv2.findContours(canny, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
cnts = cnts[0] if len(cnts) == 2 else cnts[1]

for c in cnts:
    cv2.drawContours(image,[c], 0, (0,255,0), 3)

cv2.imshow("result", image)
cv2.waitKey(0)
```
